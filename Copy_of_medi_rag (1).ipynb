```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "mistralai/Mistral-7B-Instruct-v0.2"

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_4bit=True,
    torch_dtype=torch.float16
)

model.eval()
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

pdf_path = "/content/The_Merck_Manual_of_Diagnosis_and_Therapy_2011_-_19th_Edn........pdf"

loader = PyMuPDFLoader(pdf_path)
docs = loader.load()

splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    encoding_name="cl100k_base",
    chunk_size=512,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)
print("Total chunks:", len(chunks))
```

```python
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain_community.vectorstores import Chroma
import os

embedding_model = SentenceTransformerEmbeddings(
    model_name="BAAI/bge-large-en-v1.5"
)

persist_dir = "merck_db"
os.makedirs(persist_dir, exist_ok=True)

vectorstore = Chroma.from_documents(
    chunks,
    embedding_model,
    persist_directory=persist_dir
)
```

```python
retriever = vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 6, "lambda_mult": 0.25}
)

SIMILARITY_THRESHOLD = 0.4
MAX_NEW_TOKENS = 256

```

```python
def build_prompt(context, question):
    return f"""[INST]
You are a medical assistant.
Answer strictly using the provided context.
If the answer is not present in the context, say "I don't know."

Context:
{context}

Question:
{question}
[/INST]
"""
```

```python
SIMILARITY_THRESHOLD = 0.6
MAX_NEW_TOKENS = 256

def generate_rag_response(question):

    # 1️⃣ MMR retrieval (diverse candidates)
    mmr_docs = retriever.invoke(question)
    if not mmr_docs:
        return "I don't know."

    mmr_texts = {doc.page_content for doc in mmr_docs}

    # 2️⃣ Similarity search with score (for scoring)
    scored = vectorstore.similarity_search_with_score(
        question,
        k=20
    )

    # 3️⃣ Re-rank ONLY MMR docs using distance
    reranked = [
        (doc, score)
        for doc, score in scored
        if doc.page_content in mmr_texts
    ]

    if not reranked:
        return "I don't know."

    # 4️⃣ Sort by distance (lower = better)
    reranked.sort(key=lambda x: x[1])

    # 5️⃣ Threshold guardrail
    if reranked[0][1] > SIMILARITY_THRESHOLD:
        return "I don't know."

    # 6️⃣ Build context from top-3
    context = "\n\n".join(doc.page_content for doc, _ in reranked[:3])

    # 7️⃣ Prompt + generation
    prompt = build_prompt(context, question)
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=MAX_NEW_TOKENS,
            do_sample=False
        )

    gen_tokens = outputs[0][inputs["input_ids"].shape[-1]:]
    answers= tokenizer.decode(gen_tokens, skip_special_tokens=True).strip()
    answer= answers.strip()

    return answer,context
```

```python
def generate_rag_response(question):

    # 1️⃣ MMR retrieval
    mmr_docs = retriever.invoke(question)
    if not mmr_docs:
        return None, None

    mmr_texts = {doc.page_content for doc in mmr_docs}

    # 2️⃣ Similarity search
    scored = vectorstore.similarity_search_with_score(
        question,
        k=20
    )

    # 3️⃣ Re-rank
    reranked = [
        (doc, score)
        for doc, score in scored
        if doc.page_content in mmr_texts
    ]

    if not reranked:
        return None, None

    # 4️⃣ Sort by distance
    reranked.sort(key=lambda x: x[1])

    # 5️⃣ Threshold
    if reranked[0][1] > SIMILARITY_THRESHOLD:
        return None, None

    # 6️⃣ Context
    context = "\n\n".join(doc.page_content for doc, _ in reranked[:3])

    # 7️⃣ Generate
    prompt = build_prompt(context, question)
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=MAX_NEW_TOKENS,
            do_sample=False
        )

    gen_tokens = outputs[0][inputs["input_ids"].shape[-1]:]
    answer = tokenizer.decode(gen_tokens, skip_special_tokens=True).strip()

    return answer, context
```

```python
import torch

def llm(prompt, max_tokens=128, temperature=0):
    """
    Simple LLM wrapper so evaluation code works like OpenAI-style responses.
    """

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_tokens,
            do_sample=(temperature > 0),
            temperature=temperature if temperature > 0 else None
        )

    # Remove prompt tokens
    generated_tokens = outputs[0][inputs["input_ids"].shape[-1]:]
    text = tokenizer.decode(generated_tokens, skip_special_tokens=True)

    # OpenAI-like response format
    return {
        "choices": [
            {
                "text": text.strip()
            }
        ]
    }
```

```python
groundedness_system_message = """
You are an evaluator.

Rate how well the answer is supported by the given context.
Return ONLY one number based on your confidence.
Score the answer ONLY based on how well it is supported by the given context.

Use this rubric:
5 = Fully supported, no missing details
4 = Mostly supported, minor gaps
3 = Partially supported, important details missing
2 = Weakly supported, mostly inferred
1 = Very weak support
0 = Not supported at all


Return ONLY ONE number from this set:
[0, 1, 2, 3, 4, 5]

Return ONLY the number. No explanation.

"""
```

```python
relevance_system_message = """
You are an evaluator.

Rate how well the answer addresses the user's question.
Return ONLY one number based on your confidence.
Judge ONLY relevance (not factual correctness, not grounding).

Use this rubric:

5 = Directly and completely answers the question
4 = Mostly answers the question, minor details missing
3 = Partially answers the question, important aspects missing
2 = Weakly related to the question
1 = Very little relevance
0 = Completely irrelevant

Return ONLY ONE number from this set:
[0, 1, 2, 3, 4, 5]

Return ONLY the number. No explanation.


"""
```

```python
import json
import re

def safe_json_parse(text):
    """
    Extracts and parses the first JSON object found in text.
    Returns None if parsing fails.
    """
    try:
        # Try direct parse first
        return json.loads(text)
    except json.JSONDecodeError:
        pass

    # Try to extract JSON using regex
    try:
        match = re.search(r'\{.*\}', text, re.DOTALL)
        if match:
            return json.loads(match.group())
    except Exception:
        pass

    return None
```

```python
def evaluate_answer(question, answer, context, max_tokens=10):

    # ---------- Groundedness ----------
    groundedness_prompt = f"""[INST]
{groundedness_system_message}

Context:
{context}

Question:
{question}

Answer:
{answer}
[/INST]"""

    g_response = llm(
        prompt=groundedness_prompt,
        max_tokens=max_tokens,
        temperature=0
    )

    groundedness_text = g_response["choices"][0]["text"].strip()

    groundedness_score = int(
        next((c for c in groundedness_text if c.isdigit()), -1)
    )

    # ---------- Relevance ----------
    relevance_prompt = f"""[INST]
{relevance_system_message}

Context:
{context}

Question:
{question}

Answer:
{answer}
[/INST]"""

    r_response = llm(
        prompt=relevance_prompt,
        max_tokens=max_tokens,
        temperature=0
    )

    relevance_text = r_response["choices"][0]["text"].strip()

    relevance_score = int(
        next((c for c in relevance_text if c.isdigit()), -1)
    )

    return {
        "groundedness_score": groundedness_score,
        "relevance_score": relevance_score
    }
```

```python
query = "What is the protocol for managing sepsis in a critical care unit?"

answer, context = generate_rag_response(query)

if answer is None:
    print("ANSWER: I don't know.")
else:
    evaluation = evaluate_answer(query, answer, context)

    print("ANSWER:\n", answer)
    print("GROUNDEDNESS SCORE:", evaluation["groundedness_score"])
    print("RELEVANCE SCORE:", evaluation["relevance_score"])
```
